# Seaborn

```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
```

```python
tips = sns.load_dataset('tips')
iris = sns.load_dataset('iris')
flight = sns.load_dataset('flights')

male = tips[tips['sex']=='Male']
female = tips[tips['sex']=='Female']
# print(f"Male average total bill: {male['total_bill'].mean()}\nFemale average total bill: {female['total_bill'].mean()}\nTotal Average bill: {tips['total_bill'].mean()}")

print(tips.groupby('sex')['total_bill'].mean())
print(tips.groupby('sex')['tip'].mean())

print(male.shape[0])
print(female.shape[0])
```

```python
sns.scatterplot(x='total_bill',y='tip',data=tips)
plt.show()
```

```python
print(tips.groupby('day')['total_bill'].mean())
```

```python
sns.countplot(x='day',data=tips)
```

```python
sns.countplot(x='sex',data=tips)
```

```python
sns.countplot(x='time',data=tips)
```

Load the iris dataset and make a scatterplot of sepal_length vs sepal_width.

```python
sns.scatterplot(x='sepal_length',y='petal_length',data=iris)
plt.show()
```

Using the tips dataset, make a barplot showing the average tip for each day.

```python
sns.barplot(x='day',y='tip',data=tips)
plt.show()
```

Using the flights dataset, make a countplot of year.

```python
# sns.countplot(x='year',data=flight)
# plt.show()
sns.lineplot(x='year',y='passengers',data=flight)
plt.show()
```

Histograms

```python
sns.histplot(data=tips, x='total_bill',bins=20,kde=True)
plt.show()
```

```python
sns.kdeplot(x='total_bill',fill=False,data=tips)
plt.show()
```

```python
sns.boxplot(x='day',y='total_bill',data=tips) # boxplot shows min,q1,median,q3,max then outliers
plt.show() 
```

```python
# violinplot is like bada bhai of boxplot as it is boxplot + kde(bell curve (normal dist))
sns.violinplot(x='day',y='total_bill',data=tips)
plt.show()
```

Plot a histogram of tip values with KDE overlay from the tips dataset.

```python
sns.histplot(x='tip',bins=20,data=tips,kde=True)
plt.show()
```

Compare the distribution of total_bill for each day using a boxplot.

```python
sns.boxplot(x='day',y='total_bill',data=tips)
plt.show()
```

Show a violinplot of tip vs. sex and explain what you observe.

```python
sns.violinplot(x='sex',y='tip',data=tips)
plt.show()
```
