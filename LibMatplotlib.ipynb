# Matplotlib
### The matplotlib is a powerful library for visualizing data.

```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import seaborn as sns
```

```python
x = np.arange(2,22,2)
y = np.arange(1,11)
print(x)
plt.plot(x,y)
plt.show()
```

```python
plt.plot(x,y,color="red",linestyle='--',marker='*',label="Label 1")
plt.ylabel("Y-axis")
plt.xlabel("X-axis")
plt.title("My first plot")
plt.legend()
plt.show()
```

```python
plt.bar(x,y,color="red",label="1 Bar")
plt.title("Bar Graph")
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.legend()
plt.show()
```

## Multiple Lines 

```python
x = np.arange(1,6)
y1 = np.arange(2,11,2)
y2 = np.random.randint(1,6,5)

plt.plot(x,y1,color='red',marker='o',label="line 1")
plt.plot(x,y2,color='blue',marker='^',label="line 2")
plt.title("Multi-line Graph")
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.legend()
plt.savefig("multline.png")
plt.show()
```

# Pie chart

```python
sizes = [30, 20, 25, 25]       # values (percentage-like)
labels = ["Apples", "Bananas", "Cherries", "Dates"]  
colors = ["red", "yellow", "pink", "brown"]  
explode = [0, 0.1, 0, 0]       # "explode" Bananas outwards  

plt.pie(
    sizes,
    labels=labels,          # add labels
    colors=colors,          # custom colors
    explode=explode,        # highlight one part
    autopct="%.1f%%",       # show percentages
    shadow=True,            # add shadow effect
    startangle=90           # rotate start
)
plt.title("Fruit Distribution")
plt.show()
```

```python
fail = [22,15,18,45]
classes = ['C',"Python","DBMS","DSA"]
colors = ['blue','green','yellow','red']
explode = [0,0.1,0,0.1]

plt.pie(
    fail,
    labels=classes,
    colors=colors,
    explode= explode,
    autopct='%.1f%%',
    shadow=True,
    startangle=90   
)
plt.title("Fail percentage")
plt.savefig("Piechart.png")
plt.show()
```

## Using NumPy arrays directly

```python
x = np.linspace(1,10,50)
y= np.sin(x)
z= np.cos(x)

plt.plot(x,y,color="red",marker='s',label='sin')
plt.plot(x,z,color="blue",marker = '^',label='cos')
plt.title("Sin and Cos")
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.legend()
plt.show()
```

### Using Pandas DataFrame directly

```python
data = pd.DataFrame({
    "name": ["Javed","Nika","Luffy"],
    "age": [21,1000,22],
    "fight": [3,100,20]
})

data.plot(x="name",y=['age','fight'],kind='line',marker='s')
plt.title("DataFrame plotting")
plt.xlabel("Names")
plt.ylabel("AGE AND FIGHT")
plt.show()
```

```python
data = pd.DataFrame({
    'Player': ['Virat', 'Rohit', 'Dhoni', 'Sachin', 'Kohli'],
    'Runs': [85, 120, 45, 150, 95],
    'Matches': [250, 230, 350, 463, 254],
    'StrikeRate': [92, 88, 86, 89, 93]
})

data.plot(x='Player',y='Runs',kind='bar')
plt.title("Players and Runs")
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.show()
```

```python
plt.plot(data["Matches"], data["Runs"], marker="o", linestyle="-")  
plt.xlabel("Matches")
plt.ylabel("Runs")
plt.title("Matches vs Runs")
plt.show()
```

```python
plt.scatter(data['StrikeRate'],data['Runs'],s=data['Matches'],label=data['Player'])
plt.title("StrikeRate vs Runs")
plt.legend()
plt.show()
```

# Subplots

```python
# using subplot()

x = np.arange(1,8)
y1 = np.random.randint(1,8,7)
y2 = np.random.randint(1,8,7)

plt.subplot(1,2,1)
plt.plot(x,y1,label='first')
plt.legend()

plt.subplot(1,2,2)
plt.plot(x,y2,label='second')
plt.legend()

plt.subplot(2,2,3)
plt.plot(x,x,label='third')
plt.legend()

plt.subplot(2,2,4)
plt.plot(y1,y2,label='fourth')
plt.legend()

plt.show()
```

```python
# using subplots: modern clean way

fig, axes = plt.subplots(2,1,figsize=(6,6))

axes[0].plot(x,y1,label=('first'),marker='o')
axes[0].legend()

axes[1].plot(x,y2,label='second',marker='*')
axes[1].legend()

plt.show()
```
